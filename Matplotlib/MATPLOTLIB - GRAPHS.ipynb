GRAFİK NESNESİ OLUŞTURMA-RENK-İŞARET-ÇİZGİ STİLLERİ BELİRLEME

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig = plt.figure()
data = np.random.randn(50).cumsum()
x1 = fig.add_subplot(2,2,1)
x2 = fig.add_subplot(2,2,2)
x3 = fig.add_subplot(2,2,3)
x4 = fig.add_subplot(2,2,4)
x1.plot(data,"r*")
x2.plot(data,"b--")
x3.plot(data,"ko")
x4.plot(data,"y")
```

```python
x = [1,2,3,4,5]
y = [10,11,12,13,14]
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, y, linestyle = "--", color = "b")
```

```python
plt.plot(data,"mo--")
```

```python
# eksenleri etiketlendirme
# aralık için xlim
# işaret konumu için set_xticks
# işaret etiketi için set_xticklabels
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.set_title("1.grafik")
ax1.plot(data)
ax1.set_xticks([0,15,30,45])
ax2 = fig.add_subplot(2,2,2)
ax2.set_xticklabels(["bir","iki","üç","dört"], rotation =30, fontsize="small")
ax2.set_title("2. grafik")
ax2.plot(data)
ax3= fig.add_subplot(2,2,3)
ax3.set_xlabel("aşamalar")
ax3.set_title("3.grafik")
ax3.plot(data)
ax4 = fig.add_subplot(2,2,4)
ozellikler = {"title": "4.grafik",
             "xlabel":"Adımlar",
             "ylabel":"Değerler"}
ax4.set(**ozellikler)
ax4.plot(data)
```

# ÇİZGİ GRAFİĞİ

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# style metodu ile grafik görünüşü ayarlanabilir
plt.style.use("seaborn-whitegrid")
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,100)
y = np.sin(x)
ax.plot(x,y)
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x-1),color="blue")
plt.plot(x,np.sin(x-2),color="r")
plt.plot(x,np.sin(x-3),color="0.75")
plt.plot(x,np.sin(x-4),color="#000000")
```

```python
plt.plot(x, x+1, linestyle="solid")
plt.plot(x, x+2, linestyle="dashed")
plt.plot(x, x+3, linestyle="dashdot")
plt.plot(x, x+4, linestyle="dotted")
```

```python
plt.plot(x, x+1,"-r" )
plt.plot(x, x+2,"--b")
plt.plot(x, x+3,"-.g")
plt.plot(x, x+4,":k" )
```

```python
plt.plot(x,np.sin(x),"ro")
plt.xlim(-5,20)
plt.ylim(-3,3)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,10,-1,3]) # x ekseni -1 ve 10 arası
                       # y ekseni -1 ile 3 arası değerler alabilir
```

```python
plt.plot(x,np.sin(x))
plt.title("Bir sinüs eğrisi")
plt.xlabel("x değerleri")
plt.ylabel("y değerleri")
```

```python
plt.plot(x, x**2, "--r", label = "1.grafik")
plt.plot(x, x+4, ":b", label = "2.grafik")
plt.legend()
```
